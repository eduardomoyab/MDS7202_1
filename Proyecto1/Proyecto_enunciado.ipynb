```python
class Profiler():
        
    #paso 1: definir el constructor
    def __init__(self, data): 
        """
        Inicializa la clase Profiler.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos de entrada para el an√°lisis.
        """
        self.data = data
        self.timestamp = datetime.datetime.now().strftime("%d-%m-%Y")
        self.output_dir = f"EDA_{self.timestamp}"
        os.makedirs(self.output_dir, exist_ok=True)
        self.cleaned = False
        self.scaled = False
        self.data_cruda = data.copy()

    #paso 2: definir summarise
    def summarize(self, variables=None, na_threshold=0.1, outlier_threshold=0.1, unique_threshold=200,view = False):
        """
        Genera un informe resumido del conjunto de datos.

        Par√°metros:
        variables (lista, opcional): La lista de variables a resumir.
        na_threshold (float, opcional): El umbral para valores faltantes.
        outlier_threshold (float, opcional): El umbral para detectar valores at√≠picos.
        unique_threshold (int, opcional): El umbral para la cantidad de valores √∫nicos en variables categ√≥ricas.
        view (bool, opcional): Si es True, muestra el informe resumido.
        """
        if variables is None:
            variables = self.data.columns

        summary_file = os.path.join(self.output_dir, "summary.txt")
        with open(summary_file, "w") as f:
            for var in variables:
                var_data = self.data[var]
                f.write(f"Variable: {var}\n")
                f.write(f"Tipo Variable: {var_data.dtype}\n")
                f.write(f"Valores Unicos: {var_data.nunique()}\n")
                f.write(f"Porcentaje de Valores Faltantes: {var_data.isnull().mean() * 100:.2f}%\n")

                if pd.api.types.is_numeric_dtype(var_data):
                    f.write(f"Numero de Ceros: {np.sum(var_data == 0)}\n")
                    f.write(f"N√∫mero de Valores Negativos: {np.sum(var_data < 0)}\n")
                    outliers = np.abs(var_data - var_data.mean()) > 3 * var_data.std()
                    f.write(f"N√∫mero de Outliers: {np.sum(outliers)}\n")
                    f.write(f"Min: {var_data.min()}\n")
                    f.write(f"Max: {var_data.max()}\n")
                    f.write(f"Promedio: {var_data.mean()}\n")
                    percentiles = var_data.quantile([0.25, 0.5, 0.75, 1])
                    f.write(f"Percentiles: {percentiles}\n")
                    
                    if (var_data.isna().mean() > na_threshold) or (outliers.mean() > outlier_threshold):
                        f.write("Anomal√≠a Detectada!\n")
                        if var_data.isna().mean() > na_threshold:
                            f.write(f"Tipo de Anomalia: Alto Porcentaje de Valores Faltantes\n")
                        if outliers.mean() > outlier_threshold:
                            f.write(f"Tipo de Anomalia: Alto Porcentaje de Outliers\n")

                else:
                    # Alerta de alta cardinalidad
                    if var_data.nunique() > unique_threshold:
                        f.write("Anomalia Detectada! Alto Numero de Valores Unicos\n")
                        f.write(f"Tipo de Anomalia: Alta Cardinalidad (Valores Unicos: {var_data.nunique()})\n")


                f.write("\n")
        #Si view es True, se imprime en pantalla el contenido de summary
        if view:
            with open(summary_file, "r") as f:
                print(f.read())


    def plot_vars(self, variables=None, top_categories=15):
        """
        Genera gr√°ficos para variables num√©ricas y categ√≥ricas.

        Par√°metros:
        variables (lista, opcional): La lista de variables a graficar.
        top_categories (int, opcional): N√∫mero de categor√≠as principales a mostrar para variables categ√≥ricas.
        """
        categorical_variables = []
        if variables is None:
            variables = self.data.columns

        plots_dir = os.path.join(self.output_dir, "plots")
        os.makedirs(plots_dir, exist_ok=True)

        for var in variables:
            var_data = self.data[var]
            if pd.api.types.is_numeric_dtype(var_data):
                
                # Generar grafico de densidad mediante histograma
                plt.figure(figsize=(10,8))
                sns.kdeplot(var_data, shade=True, color="r", label=var)
                plt.title(f'Densidad de distribuci√≥n de {var}')
                plt.savefig(f'{plots_dir}/{var}_density_plot.pdf')
                plt.close()


            else:
                
                #Generar histograma para las N categorias principales de variables categoricas, ordenando de mayor a menor
                if pd.api.types.is_categorical_dtype(var_data):
                    top_categories_data = var_data.value_counts().head(top_categories).sort_values(ascending=False)
                    plt.figure(figsize=(10,8))
                    top_categories_data.plot(kind='bar')
                    plt.title(f'Principales {top_categories} categorias de {var}')
                    plt.savefig(f'{plots_dir}/{var}_top_categories_histogram.pdf')
                    plt.close()
                    #Almacenamos el nombre de las variables categoricas en una lista
                    categorical_variables.append(var)
            
            #Ahora graficamos el coeficiente de Cramer para las variables categoricas
            if len(categorical_variables) > 1:
                #Generamos una matriz de correlacion entre las variables categoricas
                corr_matrix = pd.DataFrame(index=categorical_variables, columns=categorical_variables)
                for var_1 in categorical_variables:
                    for var_2 in categorical_variables:
                        corr_matrix.loc[var_1, var_2] = self.cramers_v(self.data[var_1], self.data[var_2])
                
                #Graficamos la matriz de correlacion
                plt.figure(figsize=(10,8))
                sns.heatmap(corr_matrix.astype(float), annot=True,cmap='coolwarm', fmt=".2f")
                plt.title('Coeficiente de Cramers V')
                plt.savefig(f'{plots_dir}/cramer_correlation_matrix.pdf')
                plt.close()
        #Generar matriz de correlacion con variables numericas o categoricas
        plt.figure(figsize=(12,8))

        correlation_matrix = self.data.corr(numeric_only=True)
        #Graficamos la matriz de correlacion mostrando los valores
        sns.heatmap(correlation_matrix,cmap='coolwarm', fmt=".2f",annot=True)
        plt.title('Matriz de correlaci√≥n')
        plt.savefig(f'{plots_dir}/correlation_matrix.pdf')
        plt.tight_layout() 
        plt.close()



    def cramers_v(self, x, y):
        """
        Calcula el coeficiente de asociaci√≥n Cramer's V entre dos variables categ√≥ricas.

        Par√°metros:
        x (pandas.Series): La primera variable categ√≥rica.
        y (pandas.Series): La segunda variable categ√≥rica.

        Retorna:
        float: El coeficiente de Cramer's V.
        """
        confusion_matrix = pd.crosstab(x,y)
        chi2 = chi2_contingency(confusion_matrix)[0]
        n = confusion_matrix.sum().sum()
        phi2 = chi2/n
        r,k = confusion_matrix.shape
        phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
        rcorr = r-((r-1)**2)/(n-1)
        kcorr = k-((k-1)**2)/(n-1)
        return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))


    def clean_data(self, data, variables=None, drop_duplicates=True, drop_na=True,n_cat = 60, reemplazar = False,clean_data_cruda = False):
        """
        Limpia y preprocesa el conjunto de datos.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos a limpiar.
        variables (lista, opcional): La lista de variables a considerar en la limpieza.
        drop_duplicates (bool, opcional): Si se deben eliminar las filas duplicadas.
        drop_na (bool, opcional): Si se deben eliminar los valores nulos.
        n_cat (int, opcional): N√∫mero m√°ximo de categor√≠as √∫nicas para transformar en categ√≥ricas.
        reemplazar (bool, opcional): Si se debe reemplazar el conjunto de datos original.
        clean_data_cruda (bool, opcional): Si se debe limpiar el conjunto de datos crudos original.

        Retorna:
        pd.Dataframe: Conjunto de datos limpio.
        """

        clean_data_dir = os.path.join(self.output_dir, "clean_data")
        os.makedirs(clean_data_dir, exist_ok=True)
        clean_data_file = os.path.join(clean_data_dir, "data.csv")

        if variables is None:
            variables = data.columns

        if reemplazar == False:
            cleaned_data = data[variables].copy()
        else:
            cleaned_data = data

        # fijamos como indice la columna que tenga mayor numero de valores unicos
        cleaned_data.set_index(cleaned_data.columns[cleaned_data.nunique().argmax()], inplace=True)

        #Transformamos la columna Medal
        process_medal_transformer = FunctionTransformer(self.process_medal, validate=False)
        cleaned_data = process_medal_transformer.transform(cleaned_data)

        #Separamos las columnas no atomicas
        separate_non_atomic_transformer = FunctionTransformer(self.separate_non_atomic, validate=False)
        cleaned_data = separate_non_atomic_transformer.transform(cleaned_data)

        #Guardamos los valores de drop_na
        self.drop_na = drop_na  # Almacena el valor en una variable de clase

        if drop_duplicates:
            remove_duplicates_transformer = FunctionTransformer(self.remove_duplicates, validate=False)
            cleaned_data = remove_duplicates_transformer.transform(cleaned_data)

        #Tratamos los nulos
        handle_missing_values_transformer = FunctionTransformer(self.handle_missing_values, validate=False)
        cleaned_data = handle_missing_values_transformer.transform(cleaned_data)

        #Transformamos categoricas
        self.n_cat = n_cat
        to_categorical_transformer = FunctionTransformer(self.to_categorical, validate=False)
        cleaned_data = to_categorical_transformer.transform(cleaned_data)

        cleaned_data.to_csv(clean_data_file, index=False)
        self.cleaned = True
        data = cleaned_data
        self.data = cleaned_data
        if clean_data_cruda:
            self.cleaned = False

        return cleaned_data

    #Funciones Auxiliares para el clean_data
    #-Remover duplicados
    def remove_duplicates(self, data):
        """
        Elimina filas duplicadas del conjunto de datos.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos a procesar.

        Retorna:
        pandas.DataFrame: El conjunto de datos sin filas duplicadas.
        """
        data.drop_duplicates(inplace=True)
        return data

    #-Tratamiento de valores faltantes
    def handle_missing_values(self, data):
        """
        Gestiona los valores faltantes en el conjunto de datos.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos a procesar.

        Retorna:
        pandas.DataFrame: El conjunto de datos con valores faltantes gestionados.
        """
        if self.drop_na: #Eliminamos los valores nulos
            data.dropna(inplace=True)
        else: #Impuntamos los valores nulos
            for column in data.columns:
                if pd.api.types.is_numeric_dtype(data[column]):
                    imputer = SimpleImputer(strategy='mean')
                else:
                    imputer = SimpleImputer(strategy='most_frequent')
                data[column] = imputer.fit_transform(data[column].values.reshape(-1, 1)).flatten()
        return data

    def separate_non_atomic(self, data, separator=[':', '(', '?', '*']):
        """
        Separa las columnas no at√≥micas del conjunto de datos.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos a procesar.
        separator (lista, opcional): Lista de separadores para identificar columnas no at√≥micas.

        Retorna:
        pandas.DataFrame: El conjunto de datos con columnas no at√≥micas separadas.
        """
        separator_2 = [re.escape(char) for char in separator]
        no_atomicas = []

        for var in data.columns:
            if not pd.api.types.is_numeric_dtype(data[var]) and pd.api.types.is_string_dtype(data[var]):  # Verifica si es una columna de cadena
                for sep in separator_2:
                    if data[var].str.contains(sep).any() and not data[var].str.contains("\)").any():
                        no_atomicas.append(var)
                        break

        for var in no_atomicas:
            for sep in separator:
                data[var] = data[var].str.replace(sep, '-')

            n = data[var].str.count('-').max()

            for i in range(n + 1):
                data[var + '_' + str(i)] = data[var].str.split('-').str[i]
                # Las transformamos en num√©ricas
                data[var + '_' + str(i)] = pd.to_numeric(data[var + '_' + str(i)], errors='coerce')

            data.drop(var, axis=1, inplace=True)

        data.replace('nan', np.nan, inplace=True)

        return data

    
    #-Transformacion de categoricas
    def to_categorical(self, data, variables=None):
        """
        Convierte variables con un n√∫mero limitado de valores √∫nicos a categ√≥ricas.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos a procesar.
        variables (lista, opcional): Lista de variables a convertir a categ√≥ricas.

        Retorna:
        pandas.DataFrame: El conjunto de datos con variables convertidas a categ√≥ricas.
        """
        if variables is None:
            variables = data.columns

        for var in variables:
            #Vemos si la cantidad de valores unicos es menor a 100
            if data[var].nunique() < self.n_cat:
                #Si es menor a 100, la transformamos en categorica
                data[var] = data[var].astype('category')
        return data
    
    #-Procesamiento de columna de medallas
    def process_medal(self, data):
        """
        Procesa la columna de medallas en el conjunto de datos.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos a procesar.

        Retorna:
        pandas.DataFrame: El conjunto de datos con la columna de medallas procesada.
        """
        #Vemos la columna que tenga sobre el 75% de valores nulos
        columna = data.columns[data.isnull().mean() > 0.75]
        #Transformamos los nulos de la columna a "No Medal"
        data[columna] = data[columna].fillna("No Medal")
        return data
    
    def scale(self, data, numeric_scaler=None, categorical_encoder=None):
        """
        Escala y transforma las variables num√©ricas y categ√≥ricas.

        Par√°metros:
        data (pandas.DataFrame): El conjunto de datos a escalar.
        numeric_scaler (lista, opcional): El escalador para variables num√©ricas.
        categorical_encoder (objeto, opcional): El codificador para variables categ√≥ricas.

        Retorna:
        pd.Dataframe: Conjunto de datos escalado.
        """
        scale_dir = os.path.join(self.output_dir, "scale")
        os.makedirs(scale_dir, exist_ok=True)

        numeric_features = data.select_dtypes(include=[np.number])
        categorical_features = data.select_dtypes(include=['category'])

        if numeric_scaler is None:
            numeric_scaler = [FunctionTransformer(np.log1p), MinMaxScaler()]

        numbers = numeric_features.columns.copy()
        #Variables numericas
        for col in numeric_features.columns:
            for scaler in numeric_scaler:
                transformer = ColumnTransformer(transformers=[('num', scaler, [col])])
                numeric_transformer = transformer.fit_transform(data[[col]])
                data[col] = numeric_transformer[:, 0]

        #Variables Categoricas 
        if categorical_encoder is None:
            categorical_encoder = OneHotEncoder()


        if categorical_encoder is not None:
            if isinstance(categorical_encoder, LabelEncoder):
                #LabelEncoder
                for col in categorical_features.columns:
                    encoder = LabelEncoder()
                    data[col] = encoder.fit_transform(data[col])
            elif isinstance(categorical_encoder, OneHotEncoder):
                #OneHotEncoder
                categorical_encoder.sparse_output = False
                cols_cat = categorical_features.columns.to_list()
                ct = ColumnTransformer(transformers=[('cat', categorical_encoder, cols_cat)], remainder='passthrough')
                x = ct.fit_transform(data)

                #Nuevo Df
                df = pd.DataFrame(x)
                nombres = ct.named_transformers_['cat'].get_feature_names_out(cols_cat)
                df.columns = nombres.tolist() + data.columns[~data.columns.isin(cols_cat)].tolist()
                order = data.columns[~data.columns.isin(cols_cat)].tolist() + nombres.tolist()
                df = df.reindex(columns=order)
                data = df
                #Dejamos las variables codificadas como numericas
                for col in nombres:
                    data[col] = data[col].astype('int64') 

                #Dejamos las variables numericas como numericas+
                for col in numbers:
                    data[col] = data[col].astype('float64')

        # le damos el formato de origen a las variables categoricas

        data.to_csv(os.path.join(scale_dir, "scaled_data.csv"), index=False)
        self.scaled = True
        self.data = data
        return data



    def make_clusters(self, clustering_algorithm, max_clusters=10, pca_components=2):
        """
        Realiza el an√°lisis de cl√∫steres sobre el conjunto de datos.

        Par√°metros:
        clustering_algorithm (objeto): Algoritmo de clustering a utilizar.
        max_clusters (int, opcional): El n√∫mero m√°ximo de cl√∫steres a considerar.
        pca_components (int, opcional): N√∫mero de componentes PCA para la visualizaci√≥n.

        Retorna:
        pd.Dataframe: Conjunto de datos con etiquetas de cl√∫ster.
        """
        clusters_dir = os.path.join(self.output_dir, "clusters")
        os.makedirs(clusters_dir, exist_ok=True)
        clusters_file = os.path.join(clusters_dir, "data_clusters.csv")

        data_for_clustering = self.data.copy()
        data_cruda = self.data_cruda.copy()
        data_cruda_limpia = self.clean_data(data_cruda, clean_data_cruda=True)

        #Creamos un pipeline que procese los datos con los metodos clean_data() y scale() en caso que no esten limpios y escalados

        steps_pipeline = []

        if not self.cleaned:
            steps_pipeline.append(('clean_data', FunctionTransformer(self.clean_data, validate=False)))

        if not self.scaled:
            steps_pipeline.append(('scale', FunctionTransformer(self.scale, validate=False)))


        if not self.cleaned or not self.scaled:
            pipeline = Pipeline(steps=steps_pipeline)
            data_for_clustering = pipeline.fit_transform(data_for_clustering)

        #Filtrar columnas de interes
        data_cluster_filtrada = data_for_clustering.select_dtypes(include=[np.number]).copy()

        #Visualizer para metodo del codo
        model = clustering_algorithm(random_state=10)
        visualizer = KElbowVisualizer(model, k=(2, max_clusters+1))


        visualizer.fit(data_cluster_filtrada.sample(frac=0.1, random_state=42))
        visualizer.show()
        #Obtener el numero optimo de clusters
        optimal_clusters = visualizer.elbow_value_
        print(f'Clusters seleccionados: {optimal_clusters}')

        # Ajusta el modelo de clustering con el n√∫mero √≥ptimo de clusters
        model_2 = clustering_algorithm(n_clusters=optimal_clusters,random_state=42)
        clusters = model_2.fit_predict(data_cluster_filtrada)
        data_for_clustering['cluster'] = clusters
        data_cruda_limpia['cluster'] = clusters

        # Reducci√≥n de dimensionalidad (PCA) para la visualizaci√≥n en 2D
        pca = PCA(n_components=pca_components)
        data_pca = pca.fit_transform(data_cluster_filtrada)

        # Grafica los datos en 2D y colorea por cluster con Plotly
        plt.figure(figsize=(10,8))
        fig = plt.scatter(x=data_pca[:, 0], y=data_pca[:, 1], c=data_for_clustering['cluster'],
                        alpha=1,cmap='rainbow',s=3)
        plt.title('Proyecci√≥n PCA con Etiquetas de Cluster')
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.show()

        # Guarda el n√∫mero √≥ptimo de clusters en un archivo
        optimal_clusters_file = os.path.join(clusters_dir, "optimal_clusters.txt")
        with open(optimal_clusters_file, "w") as f:
            f.write(f"N√∫mero √≥ptimo de clusters: {optimal_clusters}\n")

        # Guarda los datos con etiquetas de cluster
        data_for_clustering.to_csv(clusters_file, index=False)

        return data_cruda_limpia

    def detect_anomalies(self, anomaly_detector, pca_components=2):
        """
        Detecta anomal√≠as en el conjunto de datos.

        Par√°metros:
        anomaly_detector (objeto): Detector de anomal√≠as a utilizar.
        pca_components (int, opcional): N√∫mero de componentes PCA para la visualizaci√≥n.

        Retorna:
        pd.Dataframe: Conjunto de datos con etiquetas de anomal√≠a.
        """
        # se aplica sobre todas las variables menos las de tipo 'object'

        # Crear la carpeta 'anomalies' en la carpeta de salida
        anomalies_dir = os.path.join(self.output_dir, "anomalies")
        os.makedirs(anomalies_dir, exist_ok=True)

        raw_data = self.data.copy()

        if 'cluster' in raw_data.columns:
            raw_data = raw_data.drop(columns=['cluster'])
        
        steps_pipeline = []

        if not self.cleaned:
            steps_pipeline.append(('clean_data', FunctionTransformer(self.clean_data, validate=False)))

        if not self.scaled:
            steps_pipeline.append(('scale', FunctionTransformer(self.scale, validate=False)))
        

        if not self.cleaned or not self.scaled:
            pipeline = Pipeline(steps=steps_pipeline)
            raw_data = pipeline.fit_transform(self.data)

        # Filtrar las columnas que son num√©ricas o categ√≥ricas (transformadas en int)
        valid_columns = raw_data.select_dtypes(include=[np.number]).columns

        if not valid_columns.empty:
            numeric_data = raw_data[valid_columns]

            # Verificar si hay al menos una columna num√©rica o categ√≥rica para realizar la detecci√≥n de anomal√≠as
            if numeric_data.shape[1] > 0:
                # Aplicar el detector de anomal√≠as (anomaly_detector) a las variables num√©ricas o categ√≥ricas
                anomaly_labels = anomaly_detector.fit_predict(numeric_data)

                # Proyectar los datos a 2 dimensiones usando PCA
                pca = PCA(n_components=pca_components)
                data_pca = pca.fit_transform(numeric_data)

                # Graficar los datos en 2D y colorear por las etiquetas de anomal√≠a con Plotly
                plt.figure(figsize=(10,8))
                fig = plt.scatter(x=data_pca[:, 0], y=data_pca[:, 1], c=anomaly_labels,
                               cmap='rainbow',s=4)
                plt.title('Proyecci√≥n PCA con Etiquetas de Anomal√≠a')
                plt.xlabel('PC1')
                plt.ylabel('PC2')
                plt.show()

                # Guardar los datos con etiquetas de anomal√≠a en un archivo CSV
                numeric_data['anomaly'] = anomaly_labels
                raw_data['anomaly'] = anomaly_labels
                anomalies_file = os.path.join(anomalies_dir, "data_anomalies.csv")
                numeric_data.to_csv(anomalies_file, index=False)
            else:
                print("No hay variables num√©ricas o categ√≥ricas para la detecci√≥n de anomal√≠as.")
        else:
            print("No hay variables num√©ricas o categ√≥ricas para la detecci√≥n de anomal√≠as.")

        return raw_data


    def profile(self,clustering_algorithm, anomaly_detector, numeric_scaler = None, categorical_encoder = None,view=False):
        """
        Realiza un perfilado completo del conjunto de datos.

        Par√°metros:
        clustering_algorithm (objeto): Algoritmo de clustering a utilizar.
        anomaly_detector (objeto): Detector de anomal√≠as a utilizar.
        numeric_scaler (lista, opcional): El escalador para variables num√©ricas.
        categorical_encoder (objeto, opcional): El codificador para variables categ√≥ricas.
        view (bool, opcional): Si se debe mostrar la informaci√≥n generada.
        """
        # Paso 8: M√©todo para realizar el perfilado
        
        if not self.cleaned:
            self.clean_data(self.data)
    
        self.summarize(view=view)
        self.plot_vars()
        if not self.scaled:
            self.scale(self.data,numeric_scaler, categorical_encoder)

        # Realiza el clustering y guarda los resultados
        self.make_clusters(clustering_algorithm)
        
        # Realiza la detecci√≥n de anomal√≠as y guarda los resultados
        self.detect_anomalies(anomaly_detector)


    def clearGarbage(self):
        """
        Elimina los directorios/archivos creados por la clase Profiler.
        """
        # Paso 9: M√©todo para eliminar las carpetas/archivos creados por la clase
        shutil.rmtree(self.output_dir)
```

Con los datos ya limpios y con los escalamientos aplicados, se aplica un m√©todo de clusters a trav√©s de K-Modes, esto debido a que se tiene una baja cantidad de variables num√©ricas y alta cantidad de variables categ√≥ricas. Para determinar el n√∫mero de clusters a utilizar, se emple√≥ el m√©todo del codo de forma √≥ptima, considerando un rango de 2 a 10 clusters, esto debido a que 2 es el valor m√≠nimo de informaci√≥n que aportar√≠a en los datos, y 10 debido a que se posee una gran cantidad de datos, por ende se debe ser consciente del costo computacional que esto puede implicar al momento de hacer el an√°lisis de codo.
