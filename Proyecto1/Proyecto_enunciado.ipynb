```python
class Profiler():
        
    #paso 1: definir el constructor
    def __init__(self, data): 
        self.data = data
        self.timestamp = datetime.datetime.now().strftime("%d-%m-%Y")
        self.output_dir = f"EDA_{self.timestamp}"
        os.makedirs(self.output_dir, exist_ok=True)

    #paso 2: definir summarise
    def summarize(self, variables=None, na_threshold=0.1, outlier_threshold=0.1, unique_threshold=201):
        if variables is None:
            variables = self.data.columns

        summary_file = os.path.join(self.output_dir, "summary.txt")
        with open(summary_file, "w") as f:
            for var in variables:
                var_data = self.data[var]
                f.write(f"Variable: {var}\n")
                f.write(f"Variable Type: {var_data.dtype}\n")
                f.write(f"Number of Unique Values: {var_data.nunique()}\n")
                f.write(f"Percentage of Missing Values: {var_data.isnull().mean() * 100:.2f}%\n")

                if pd.api.types.is_numeric_dtype(var_data):
                    f.write(f"Number of Zeros: {np.sum(var_data == 0)}\n")
                    f.write(f"Number of Negative Values: {np.sum(var_data < 0)}\n")
                    outliers = np.abs(var_data - var_data.mean()) > 3 * var_data.std()
                    f.write(f"Number of Outliers: {np.sum(outliers)}\n")
                    f.write(f"Min: {var_data.min()}\n")
                    f.write(f"Max: {var_data.max()}\n")
                    f.write(f"Mean: {var_data.mean()}\n")
                    percentiles = var_data.quantile([0.25, 0.5, 0.75, 1])
                    f.write(f"Percentiles: {percentiles}\n")

                    # Alert for anomalies based on the defined thresholds
                    if (var_data.isna().mean() > na_threshold) or (outliers.mean() > outlier_threshold):
                        f.write("Anomaly Detected!\n")
                        if var_data.isna().mean() > na_threshold:
                            f.write(f"Anomaly Type: High Percentage of Missing Values\n")
                        if outliers.mean() > outlier_threshold:
                            f.write(f"Anomaly Type: High Percentage of Outliers\n")

                    # Add more numeric statistics here
                else:
                    # Alert for high cardinality (number of unique values) variables
                    if var_data.nunique() > unique_threshold:
                        f.write("Anomaly Detected! High Cardinality Variable\n")
                        f.write(f"Anomaly Type: High Cardinality (Unique Values: {var_data.nunique()})\n")

                    # Add more categorical statistics here

                f.write("\n")

    def plot_vars(self, variables=None, top_categories=15):
        if variables is None:
            variables = self.data.columns

        plots_dir = os.path.join(self.output_dir, "plots")
        os.makedirs(plots_dir, exist_ok=True)

        for var in variables:
            var_data = self.data[var]
            if pd.api.types.is_numeric_dtype(var_data):
                
                # Generate density plot for numeric variables
                fig = px.histogram(self.data, x=var, nbins=30, title=f'Density Plot of {var}')
                fig.update_layout(xaxis_title=var, yaxis_title='Density')
                plot_filename = os.path.join(plots_dir, f'{var}_density_plot.html')
                fig.write_html(plot_filename)

                # Generate correlation matrix plot con las variables numericas o categoricas
                corr_matrix = self.data.corr(numeric_only=True)
                fig = go.Figure(data=go.Heatmap(z=corr_matrix.values, x=corr_matrix.columns, y=corr_matrix.columns))
                fig.update_layout(title='Correlation Matrix', autosize=False)
                plot_filename = os.path.join(plots_dir, 'correlation_matrix.html')
                fig.write_html(plot_filename)
            else:
                
                # Generate histogram for the top N categories of categorical variables, ordenando de mayor a menor
                if var_data.nunique() > top_categories and pd.api.types.is_categorical_dtype(var_data):
                    top_categories_data = var_data.value_counts().head(top_categories).sort_values(ascending=False)
                    fig = px.bar(x=top_categories_data.index, y=top_categories_data.values, title=f'Top {top_categories} Categories of {var}')
                    fig.update_layout(xaxis_title='Categories', yaxis_title='Count')
                    fig.update_xaxes(categoryorder="total descending")
                    plot_filename = os.path.join(plots_dir, f'{var}_top_categories_histogram.html')
                    fig.write_html(plot_filename)

                    # Generate Cramer's V matrix for categorical variables
                    le = LabelEncoder()
                    encoded_var_data = le.fit_transform(var_data)
                    cramers_v = np.zeros((len(variables), len(variables)))

                    for i in range(len(variables)):
                        for j in range(len(variables)):
                            if pd.api.types.is_categorical_dtype(self.data[variables[j]]):
                                confusion_matrix = pd.crosstab(encoded_var_data, encoded_var_data[j])
                                chi2 = chi2_contingency(confusion_matrix)[0]
                                n = len(self.data)
                                min_dim = min(confusion_matrix.shape) - 1
                                if min_dim != 0:
                                    cramers_v[i][j] = np.sqrt(chi2 / (n * min_dim))
                                else:
                                    cramers_v[i][j] = 0

                    fig = go.Figure(data=go.Heatmap(z=cramers_v, x=variables, y=variables))
                    fig.update_layout(title="Cramer's V Matrix", autosize=False)
                    plot_filename = os.path.join(plots_dir, "cramers_v_matrix.html")
                    fig.write_html(plot_filename)


    def clean_data(self, variables=None, drop_duplicates=True, drop_na=True,n_cat = 250, reemplazar = False):
        clean_data_dir = os.path.join(self.output_dir, "clean_data")
        os.makedirs(clean_data_dir, exist_ok=True)
        clean_data_file = os.path.join(clean_data_dir, "data.csv")

        if variables is None:
            variables = self.data.columns

        if reemplazar == False:
            cleaned_data = self.data[variables].copy()
        else:
            cleaned_data = self.data

        #Transformamos la columna Medal
        process_medal_transformer = FunctionTransformer(self.process_medal, validate=False)
        cleaned_data = process_medal_transformer.transform(cleaned_data)

        #Separamos las columnas no atomicas
        separate_non_atomic_transformer = FunctionTransformer(self.separate_non_atomic, validate=False)
        cleaned_data = separate_non_atomic_transformer.transform(cleaned_data)

        #Guardamos los valores de drop_na e impute_na
        self.drop_na = drop_na  # Almacena el valor en una variable de clase

        if drop_duplicates:
            remove_duplicates_transformer = FunctionTransformer(self.remove_duplicates, validate=False)
            cleaned_data = remove_duplicates_transformer.transform(cleaned_data)

        #Tratamos los nulos
        handle_missing_values_transformer = FunctionTransformer(self.handle_missing_values, validate=False)
        cleaned_data = handle_missing_values_transformer.transform(cleaned_data)

        #Transformamos categoricas
        self.n_cat = n_cat
        to_categorical_transformer = FunctionTransformer(self.to_categorical, validate=False)
        cleaned_data = to_categorical_transformer.transform(cleaned_data)

        cleaned_data.to_csv(clean_data_file, index=False)

        return cleaned_data

    #Auxiliares para el clean_data
    def remove_duplicates(self, data):
        data.drop_duplicates(inplace=True)
        return data

    def handle_missing_values(self, data):
        if self.drop_na:
            data.dropna(inplace=True)
        else:
            for column in data.columns:
                if pd.api.types.is_numeric_dtype(data[column]):
                    imputer = SimpleImputer(strategy='mean')
                else:
                    imputer = SimpleImputer(strategy='most_frequent')
                data[column] = imputer.fit_transform(data[column].values.reshape(-1, 1)).flatten()
        return data

    def separate_non_atomic(self, data, separator=[':', "(", "?", '*']):
        separator_2 = [re.escape(char) for char in separator]
        no_atomicas = []
        
        for var in data.columns:
            if not pd.api.types.is_numeric_dtype(data[var]):
                for sep in separator_2:
                    if data[var].str.contains(sep).any() and not data[var].str.contains("\)").any():
                        no_atomicas.append(var)
                        break
        
        for var in no_atomicas:
            for sep in separator:
                data[var] = data[var].str.replace(sep, '-')
            
            n = data[var].str.count('-').max()
            
            for i in range(n+1):
                data[var + '_' + str(i)] = data[var].str.split('-').str[i]
                #Las tranformamos en numericas
                data[var + '_' + str(i)] = pd.to_numeric(data[var + '_' + str(i)], errors='coerce')
            
            data.drop(var, axis=1, inplace=True)
        
        data.replace('nan', np.nan, inplace=True)
        
        return data
    
    def to_categorical(self, data, variables=None):
        if variables is None:
            variables = data.columns

        for var in variables:
            #Vemos si la cantidad de valores unicos es menor a 100
            if data[var].nunique() < self.n_cat:
                #Si es menor a 100, la transformamos en categorica
                data[var] = data[var].astype('category')
        return data

    def process_medal(self, data):
        #Vemos la columna que tenga sobre el 75% de valores nulos
        columna = data.columns[data.isnull().mean() > 0.75]
        #Transformamos los nulos de la columna a "No Medal"
        data[columna] = data[columna].fillna("No Medal")
        return data
```
