**2. La función `describir(datos)` que generalice la función `generar_descripcion` para recibir una lista de frutas y devolver una lista con las descripciones de cada una:**

Por ejemplo, `describir([limon, naranja]])` deberá devolver una lista de la forma `[descripcion_limon, descripcion_naranja]`

````python
def describir(datos):
    """
    Genera una lista de descripciones para una lista de frutas.

    Parameters
    ----------
    datos : list
        Una lista de diccionarios donde cada diccionario contiene la información de una fruta.
        

    Returns
    -------
    descripciones : list
        Una lista de descripciones de frutas en base a su nombre, tipo, color y la presencia o ausencia de pepas.
    """
    descripciones = []

    for fruta in datos:
        descripcion = generar_descripcion(fruta)
        descripciones.append(descripcion)

    return descripciones
````

**3. La función `filtrar_por_pepa(datos, tiene_pepas)` que dado un tipo de fruta y un booleano, retorne las frutas que tienen o no tienen pepas (según el valor de `tiene_pepas`).**
    

Por ejemplo, `filtrar_por_pepa(datos, True)` deberá retornar una lista con los diccionarios de: naranja, limon, frutilla y frambuesa.

````python
def filtrar_por_pepa(datos, tiene_pepas):
    """
    Filtra una lista de frutas según si tienen o no pepas.

    Parameters
    ----------
    datos : list
        Una lista de diccionarios donde cada diccionario contiene la información de una fruta.

    tiene_pepas : bool
        Un valor booleano que indica si se desean frutas con pepas (True) o sin pepas (False).

    Returns
    -------
    filtrados : list
        Una lista de diccionarios de frutas que cumplen con el criterio de tener o no tener pepas.
    """
    filtrados = []

    for fruta in datos:
        if fruta["pepas"] == tiene_pepas:
            filtrados.append(fruta)

    return filtrados
````

**4. La función `conteo_colores(datos)` que cree un diccionario que haga un conteo los colores de las frutas.**

Para estos datos, la función debería retornar el siguiente diccionario: 
    
```python
{"amarillo": 3, "naranjo": 1, "rojo": 2}
```

````python
def conteo_colores(datos):
    """
    Cuenta la cantidad de frutas por cada color y devuelve un diccionario de conteo.

    Parameters
    ----------
    datos : list
        Una lista de diccionarios donde cada diccionario contiene la información de una fruta.
        Cada diccionario debe contener la clave "color" que indica el color de la fruta.

    Returns
    -------
    conteo : dict
        Un diccionario donde las claves son los colores de las frutas y los valores son la cantidad de frutas de cada color.
    """
    conteo = {}

    for fruta in datos:
        color = fruta["color"]
        if color in conteo:
            conteo[color] += 1
        else:
            conteo[color] = 1
    return conteo
````

**5. La función `obtener_tipos` que devuelva una lista con los tipos únicos de fruta usando `set()`**.

Por ejemplo, `obtener_tipos([limon, naranja, platano])` debería devolver `["citrica", "tropical"]`

````python
def obtener_tipos(datos):
    """
    Obtiene una lista con tipos únicos de frutas preservando el orden.

    Parameters
    ----------
    datos : list
        Una lista de diccionarios donde cada diccionario contiene la información de una fruta.
        Cada diccionario debe contener la clave "tipo" que indica el tipo de la fruta.

    Returns
    -------
    list
        Una lista con tipos únicos de frutas preservando el orden de entrada de los datos.
    """
    tipos_set = set()
    tipos = []

    for fruta in datos:
        tipo = fruta["tipo"]
        if tipo not in tipos_set:
            tipos_set.add(tipo)
            tipos.append(tipo)

    return tipos
````

**7. Programe nuevamente las funciones `describir`, `filtrar_por_pepa` y `obtener_tipos`, pero esta vez usando *comprehensions*.**

*Hint: Pueden reutilizar la documentación escrita en la pregunta 5*

````python
def obtener_descripciones_v2(datos):
    """
    Genera una lista de descripciones para una lista de frutas utilizando comprehensions.

    Parameters
    ----------
    datos : list
        Una lista de diccionarios donde cada diccionario contiene la información de una fruta.
        Cada diccionario debe contener las siguientes claves:
        - "nombre" (str): El nombre de la fruta.
        - "tipo" (str): El tipo de la fruta.
        - "color" (str): El color de la fruta.
        - "pepas" (bool): True si la fruta tiene pepas, False en caso contrario.

    Returns
    -------
    list
        Una lista de descripciones de frutas en base a su nombre, tipo, color y la presencia o ausencia de pepas.
    """
    descripciones = [generar_descripcion(fruta) for fruta in datos]
    return descripciones

def filtrar_por_pepa_v2(datos, tiene_pepas):
    """
    Filtra una lista de frutas según si tienen o no pepas.

    Parameters
    ----------
    datos : list
        Una lista de diccionarios donde cada diccionario contiene la información de una fruta.

    tiene_pepas : bool
        Un valor booleano que indica si se desean frutas con pepas (True) o sin pepas (False).

    Returns
    -------
    filtrados : list
        Una lista de diccionarios de frutas que cumplen con el criterio de tener o no tener pepas.
    """
    
    filtrados = [fruta for fruta in datos if fruta["pepas"] == tiene_pepas]
    return filtrados

def obtener_tipos_v2(datos):
    """
    Obtiene una lista con tipos únicos de frutas preservando el orden.

    Parameters
    ----------
    datos : list
        Una lista de diccionarios donde cada diccionario contiene la información de una fruta.
        Cada diccionario debe contener la clave "tipo" que indica el tipo de la fruta.

    Returns
    -------
    list
        Una lista con tipos únicos de frutas preservando el orden de entrada de los datos.
    """
    tipos_set = set()
    tipos = [tipo for fruta in datos if (tipo := fruta["tipo"]) not in tipos_set and not tipos_set.add(tipo)]

    return tipos
````

    1. La fruta sea diccionario.
    2. El diccionario entregado tenga las llaves `nombre`, `tipo`, `color` y `pepas`.

````python
class Jugera(Electrodomestico):
    
    def __init__(self):
        super().__init__()
        self._bandeja = []
    
    @property
    def bandeja(self):
        return self._bandeja
    
    @bandeja.setter
    def bandeja(self, nueva_bandeja):
        if not isinstance(nueva_bandeja, list):
            raise Exception("La bandeja debe ser una lista.")
        if len(nueva_bandeja) > 3:
            raise Exception("No se permiten más de 3 ingredientes en la bandeja.")
        for fruta in nueva_bandeja:
            if not isinstance(fruta, dict):
                raise Exception("Cada elemento de la bandeja debe ser un diccionario de fruta.")
            required_keys = {"nombre", "tipo", "color", "pepas"}
            if not all(key in fruta for key in required_keys):
                raise Exception("Cada fruta debe tener las llaves 'nombre', 'tipo', 'color' y 'pepas'.")
        self._bandeja = nueva_bandeja
    
    def agregar_ingrediente(self, nueva_fruta):
        self._bandeja.append(nueva_fruta)
     
    def listar_ingredientes(self):
        if not self._bandeja:
            print("Bandeja vacía")
        else:
            nombres_frutas = [fruta["nombre"] for fruta in self._bandeja]
            lista_ingredientes = ", ".join(nombres_frutas)
            print(f"Ingredientes en la bandeja: {lista_ingredientes}.")
    
    def preparar_jugo(self):
        self.esta_enchufado()
        if not self._bandeja:
            raise Exception("Error ❌: La bandeja no tiene ingredientes.")
        for fruta in self._bandeja:
            if fruta["pepas"]:
                print("Alerta ⚠️: El jugo puede contener restos de pepas.")
                break
        nombres_frutas = [fruta["nombre"] for fruta in self._bandeja]
        mensaje = f"Jugo de {', '.join(nombres_frutas)} listo. 🏖️🥤 Que lo disfrutes!!! 🥤🏖️."
        self._bandeja = []
        return mensaje
````
