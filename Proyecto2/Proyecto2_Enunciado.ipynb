# Preparación de los Datos


## Preprocesamiento con `ColumnTransformer`

En base a la exploración anterior, se transforman las variables categóricas y se extraen las variables numéricas.

```python
cat = ['occupation','payment_of_min_amount','payment_behaviour']
not_used = ['customer_id','monthly_inhand_salary','credit_history_age','num_of_delayed_payment']
target = 'credit_score'
num = [x for x in df.columns if x not in cat and x not in not_used and x != target]
```

Convertimos las columnas numéricas en caso que sean mal leidas.

```python
#Convertimos las variables numericas en caso que sean mal leidas
for col in num:
    df[col] = pd.to_numeric(df[col], errors='coerce')

#Convertimos las variables categoricas en caso que sean mal leidas
for col in cat:
    df[col] = df[col].astype('category')
```

Adicionalmente se realiza un tratamiendo de datos debido a outliers detectados en el análisis anterior. En primer lugar, los valores negativos seran transformados a valores positivos en las variables:
* age
* num_of_delayed_payment

Adicionalmente, la variable age sera limitada a valores entre 15 y 115 (Implementado en forma de función mas adelante)

Ahora generamos un ColumnTransformer que procese los datos categoricos y numéricos. Para ello se utiliza OneHotEncoder en el caso de las variables categóricas. Además se aplica un escalamiento de los datos numéricos utilizando RobustScaler, debido a la alta cantidad de outliers y a la distribución no normal de los datos.

```python
ct = ColumnTransformer([
    ('Categoricas', OneHotEncoder(sparse=False, handle_unknown="ignore",drop='if_binary'), cat),
    ('Numericas', RobustScaler(), num)
],
    remainder='passthrough')
ct.set_output(transform="pandas")
```

## Datos Nulos

Inicialmente se considerará como útil para el modelo toda variable que posea un porcentaje menor al 5% de datos faltantes. Las variables que cumplan este requisito, se les realizará una imputación de los datos faltantes, mientras que las variables que no cumplan el requisito, no serán consideradas para el modelo.

En ese sentido, se tiene que las variables: 
* monthly_inhand_salary (15.33%) 
* credit_history_age (8.96%) 
* num_of_delayed_payment (6.72%) 

No serán considerados para el modelo.

```python
#Variables que no serán consideradas en el modelo
not_used = ['customer_id','monthly_inhand_salary','credit_history_age','num_of_delayed_payment']

#Variables con datos nulos a imputar
to_impute = ['amount_invested_monthly','monthly_balance','num_credit_inquiries','changed_credit_limit']
```

Para realizar la imputación, se utilizará un SimpleImputer. Como se trata de datos numéricos, se utilizará el promedio de cada variable para la imputación. Realizamos este procesamiento en conjunto con el anterior en una sola función:

```python
def procesar_data(df2):
    df = df2.copy()
    cat = ['occupation','payment_of_min_amount','payment_behaviour']
    not_used = ['customer_id','monthly_inhand_salary','credit_history_age','num_of_delayed_payment']
    target = 'credit_score'
    num = [x for x in df.columns if x not in cat and x not in not_used and x != target]
        
    #Convertimos las variables numericas en caso que sean mal leidas
    for col in num:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    #Convertimos las variables categoricas en caso que sean mal leidas
    for col in cat:
        df[col] = df[col].astype('category')

    #Transformamos a valores positivos las variables age y num_of_delayed_payment
    df['age'] = df['age'].abs()
    df['age'] = df['age'].astype('int')
    df['num_of_delayed_payment'] = df['num_of_delayed_payment'].abs()

    #Limitamos la variable age entre 15 y 115
    df['age'] = df['age'].apply(lambda x: max(15, min(115, x)))

    #Eliminamos columnas que no serán utilizadas
    for col in not_used:
        if col in df.columns:
            df = df.drop(columns=col)

    #Variables con datos nulos a imputar
    to_impute = ['amount_invested_monthly','monthly_balance','num_credit_inquiries','changed_credit_limit']

    #Imputamos los datos nulos con la media
    imputer = SimpleImputer(strategy='mean')
    df[to_impute] = imputer.fit_transform(df[to_impute])

    return df
```

```python
Preprocess = FunctionTransformer(procesar_data)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import optuna
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
import time
import warnings
warnings.filterwarnings('ignore')
# Desactivar warnings de LightGBM
warnings.filterwarnings("ignore", category=UserWarning, module="lightgbm")

# Definir la función de optimización con límite de tiempo y sin verbose en LightGBM
def objective(trial, model, Preprocess=Preprocess, ct=ct):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'subsample': trial.suggest_loguniform('subsample', 0.5, 1),
    }

    model.set_params(**params) 

    pipeline = ImbPipeline(steps=[
        ("preprocessing", Preprocess),
        ("ct", ct),
        ('balancing', SMOTE(random_state=seed)),
        ("model", model)
    ])

    pipeline.fit(X_train, y_train)

    y_pred = pipeline.predict(X_test)
    precision = precision_score(y_test, y_pred)

    return precision
```

```python
# Crear y ejecutar el estudio de optimización para LightGBM
study_lgbm = optuna.create_study(direction='maximize')
model = LGBMClassifier(verbose=-1)  # Instancia del modelo

study_lgbm.optimize(lambda trial: objective(trial, model), timeout = 1*60, n_jobs=3)

print()
print()
print("Mejores parámetros para LightGBM:")
for key, value in study_lgbm.best_params.items():
    print(f"{key}: {value}")
```

```python
# Crear y ejecutar el estudio de optimización para LightGBM
study_xgb = optuna.create_study(direction='maximize')
model = XGBClassifier(random_state=seed)  # Instancia del modelo

study_xgb.optimize(lambda trial: objective(trial, model), timeout = 1*60, n_jobs=1)

print()
print()
print("Mejores parámetros para XGB:")
for key, value in study_xgb.best_params.items():
    print(f"{key}: {value}")
```

```python
def create_pipeline_balanced(DataProcessor,ColTransformer, model):
    return ImbPipeline(steps=[
        ("preprocessing", DataProcessor),
        ("ct", ColTransformer),
        ('balancing', SMOTE(random_state=seed)),
        ("model", model)
    ])
```

# Interpretabilidad

En esta etapa del análisis, nos enfocamos en explorar y comprender la interpretabilidad de un modelo de Machine Learning. Para ello, empleamos técnicas como la importancia de características, SHAP (SHapley Additive exPlanations) y Partial Dependence Plots (PDP), las cuales nos permiten explicar las decisiones tomadas por el modelo y comprender su funcionamiento interno.

La importancia de características evalúa la contribución de cada atributo a la precisión del modelo. Este método ofrece una visión general de los atributos más relevantes para el modelo, basándose en la ganancia de información o la frecuencia con que se divide una característica en modelos basados en árboles. Sin embargo, puede presentar sesgos si existen correlaciones entre variables independientes, atribuyendo importancias desproporcionadas a características correlacionadas.

Los SHAP Values proporcionan explicaciones detalladas sobre predicciones individuales, mostrando cómo cada característica contribuyó a una predicción específica. Aunque son útiles para el análisis detallado de casos, su complejidad puede dificultar su interpretación y comunicación a audiencias no técnicas.

Por otro lado, los Partial Dependence Plots (PDP) representan cómo cambia la precisión del modelo al variar una característica específica, manteniendo las demás constantes. Esta técnica permite visualizar de forma intuitiva las relaciones entre variables, mostrando cómo una característica afecta las predicciones del modelo.

La combinación de la importancia de características, SHAP y PDP brinda un análisis completo y sólido de la interpretabilidad del modelo. Mientras que la importancia de características ofrece una visión general, SHAP proporciona explicaciones detalladas para predicciones individuales, y los PDP permiten explorar las relaciones entre variables. Esta combinación enriquece nuestra comprensión del comportamiento del modelo en diferentes niveles de detalle y nos ayuda a tomar decisiones informadas sobre su utilización.

### SHAP

```python
import shap
import warnings
warnings.filterwarnings("ignore", message=".*The 'nopython' keyword.*")

#Para obtener los nombres
feature_names = ct.get_feature_names_out()
#Aplicamos el procesamiento al conjunto X
X = df.drop(columns=['credit_score'])

# Preprocesamiento de X_test
X_processed = Preprocess.transform(X)
X_processed =  pd.DataFrame(ct.transform(X_processed), columns= feature_names)

# Crear un Explainer con el modelo y los datos preprocesados
best_model = best_pipeline.named_steps['model']
explainer = shap.Explainer(best_model)

# Calcular SHAP values para el conjunto de prueba
shap_values = explainer.shap_values(X_processed)

# Crear un Explainer con el modelo y los datos preprocesados
best_model = best_pipeline.named_steps['model']
explainer = shap.Explainer(best_model)
```

```python
# Seleccionar 10 observaciones aleatorias
np.random.seed(seed)  # Para reproducibilidad
random_indices = np.random.choice(X_processed.shape[0], 10, replace=False)

# Calcular los valores SHAP para estas observaciones
shap_values_subset = explainer.shap_values(X_processed.iloc[random_indices])

shap.initjs()

# Máximo número de características a mostrar
max_display_features = 7

# Generar y mostrar los waterfall plots para cada observación
for i in range(10):
    # Crear un objeto SHAP Explanation para la observación actual
    explanation = shap.Explanation(
        values=shap_values_subset[1][i],  # Asumiendo que estás interesado en la clase 1
        base_values=explainer.expected_value[1],
        data=X_processed.iloc[random_indices[i]],
        feature_names=feature_names
    )

    # Generar el waterfall plot para la instancia actual
    shap.plots.waterfall(explanation, max_display=max_display_features)
```

Los gráficos de waterfall plot reflejan una tendencia consistente en la mayoría de las observaciones analizadas. Es evidente que la característica "outstanding_debt" tiene un impacto significativo en los resultados. Se observa que cuando el valor de esta característica es positivo, su influencia también es positiva en el "Credit Score", lo que sugiere un aumento en la probabilidad de que la variable tome el valor 1 en la predicción del modelo. Además, se destaca que la variable "interest_rate" también muestra una relevancia notable en el modelo y su capacidad predictiva. Estos hallazgos son congruentes con la expectativa, ya que estas variables tienen un sentido intuitivo en su impacto sobre el riesgo crediticio.

### Partial Dependence

```python
from sklearn.inspection import PartialDependenceDisplay

# Calcular la media de los valores absolutos de SHAP por característica
shap_sum = np.abs(shap_values[1]).mean(axis=0)
top_features = np.argsort(shap_sum)[-5:]  # Índices de las 5 características principales

# Elegir una submuestra aleatoria de 1000 observaciones de X_processed
np.random.seed(seed)  # Para reproducibilidad
subsample_indices = np.random.choice(X_processed.shape[0], 1000, replace=False)
X_subsample = X_processed.iloc[subsample_indices]

# Generar los Partial Dependence Plots para las 5 características más importantes
for feature in top_features:
    fig, ax = plt.subplots(figsize=(8, 4))
    PartialDependenceDisplay.from_estimator(
        best_model, X_subsample, [feature], ax=ax, grid_resolution=50, kind='average'
    )
    ax.set_title(f"Partial Dependence Plot para {feature}", fontsize=16)
    ax.set_xlabel(feature, fontsize=12, rotation=45)
    ax.set_ylabel("Predicted probability change", fontsize=12)
    ax.tick_params(axis='both', which='major', labelsize=10)
    plt.show()
```

Los gráficos de Partial Dependence Plots revelan cómo varía la predicción del modelo al modificar una característica específica, manteniendo todas las demás constantes. Los resultados son los siguientes:

* **Num_of_loan** (Numero de préstamos): Se observa un ascenso abrupto en cero, seguido de un descenso a un punto medio, manteniéndose constante para valores de num_of_loan mayores o iguales a 25. Esta transición sugiere que el modelo otorga poca importancia a valores bajos de num_of_loan, pero luego considera esta característica de manera más uniforme para valores más altos.

* **Num_credit_card** (Número de tarjetas de crédito): Hay una notable disminución en la dependencia parcial a medida que el número de tarjetas de crédito aumenta. Esto sugiere que tener un mayor número de tarjetas se asocia con un menor riesgo, posiblemente debido a un historial crediticio sólido o mayor solvencia.

* **Delay_from_due_date** (Retraso desde la fecha de vencimiento): Se observa una fluctuación en la dependencia parcial con un aumento general, que se eleva significativamente después de cierto punto. Esto sugiere que los retrasos más prolongados podrían estar asociados con un mayor riesgo de incumplimiento.

* **Interest_rate** (Tasa de interés): La dependencia parcial muestra un aumento notable en ciertos rangos de la tasa de interés, seguido de una estabilización. Esto indica que el modelo es sensible a cambios en la tasa de interés hasta cierto punto, después del cual los cambios adicionales tienen un impacto limitado en la predicción.

* **Outstanding_debt** (Deuda pendiente): Se evidencia un aumento en la dependencia parcial con el incremento de la deuda pendiente hasta cierto valor, después del cual la dependencia se estabiliza. Esto puede reflejar que un aumento en la deuda pendiente hasta cierto nivel incrementa el riesgo de incumplimiento, más allá del cual el riesgo no aumenta significativamente.
