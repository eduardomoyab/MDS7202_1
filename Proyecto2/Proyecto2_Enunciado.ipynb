# Preparación de los Datos


## Preprocesamiento con `ColumnTransformer`

En base a la exploración anterior, se transforman las variables categóricas y se extraen las variables numéricas.

```python
cat = ['occupation','payment_of_min_amount','payment_behaviour']
not_used = ['customer_id','monthly_inhand_salary','credit_history_age','num_of_delayed_payment']
target = 'credit_score'
num = [x for x in df.columns if x not in cat and x not in not_used and x != target]
```

Convertimos las columnas numéricas en caso que sean mal leidas.

```python
#Convertimos las variables numericas en caso que sean mal leidas
for col in num:
    df[col] = pd.to_numeric(df[col], errors='coerce')

#Convertimos las variables categoricas en caso que sean mal leidas
for col in cat:
    df[col] = df[col].astype('category')
```

Adicionalmente se realiza un tratamiendo de datos debido a outliers detectados en el análisis anterior. En primer lugar, los valores negativos seran transformados a valores positivos en las variables:
* age
* num_of_delayed_payment

Adicionalmente, la variable age sera limitada a valores entre 15 y 115 (Implementado en forma de función mas adelante)

Ahora generamos un ColumnTransformer que procese los datos categoricos y numéricos. Para ello se utiliza OneHotEncoder en el caso de las variables categóricas. Además se aplica un escalamiento de los datos numéricos utilizando RobustScaler, debido a la alta cantidad de outliers y a la distribución no normal de los datos.

```python
ct = ColumnTransformer([
    ('Categoricas', OneHotEncoder(sparse=False, handle_unknown="ignore",drop='if_binary'), cat),
    ('Numericas', RobustScaler(), num)
],
    remainder='passthrough')
ct.set_output(transform="pandas")
```

## Datos Nulos

Inicialmente se considerará como útil para el modelo toda variable que posea un porcentaje menor al 5% de datos faltantes. Las variables que cumplan este requisito, se les realizará una imputación de los datos faltantes, mientras que las variables que no cumplan el requisito, no serán consideradas para el modelo.

En ese sentido, se tiene que las variables: 
* monthly_inhand_salary (15.33%) 
* credit_history_age (8.96%) 
* num_of_delayed_payment (6.72%) 

No serán considerados para el modelo.

```python
#Variables que no serán consideradas en el modelo
not_used = ['customer_id','monthly_inhand_salary','credit_history_age','num_of_delayed_payment']

#Variables con datos nulos a imputar
to_impute = ['amount_invested_monthly','monthly_balance','num_credit_inquiries','changed_credit_limit']
```

Para realizar la imputación, se utilizará un SimpleImputer. Como se trata de datos numéricos, se utilizará el promedio de cada variable para la imputación. Realizamos este procesamiento en conjunto con el anterior en una sola función:

```python
def procesar_data(df2):
    df = df2.copy()
    cat = ['occupation','payment_of_min_amount','payment_behaviour']
    not_used = ['customer_id','monthly_inhand_salary','credit_history_age','num_of_delayed_payment']
    target = 'credit_score'
    num = [x for x in df.columns if x not in cat and x not in not_used and x != target]
        
    #Convertimos las variables numericas en caso que sean mal leidas
    for col in num:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    #Convertimos las variables categoricas en caso que sean mal leidas
    for col in cat:
        df[col] = df[col].astype('category')

    #Transformamos a valores positivos las variables age y num_of_delayed_payment
    df['age'] = df['age'].abs()
    df['age'] = df['age'].astype('int')
    df['num_of_delayed_payment'] = df['num_of_delayed_payment'].abs()

    #Limitamos la variable age entre 15 y 115
    df['age'] = df['age'].apply(lambda x: max(15, min(115, x)))

    #Eliminamos columnas que no serán utilizadas
    for col in not_used:
        if col in df.columns:
            df = df.drop(columns=col)

    #Variables con datos nulos a imputar
    to_impute = ['amount_invested_monthly','monthly_balance','num_credit_inquiries','changed_credit_limit']

    #Imputamos los datos nulos con la media
    imputer = SimpleImputer(strategy='mean')
    df[to_impute] = imputer.fit_transform(df[to_impute])

    return df
```

```python
Preprocess = FunctionTransformer(procesar_data)
```
